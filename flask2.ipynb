```python
# Q2. Why is request used in Flask?
# Ans2 Flask is a web framework for Python that allows building web applications.
# The request object in Flask is used to access the data submitted by the client to the server
# via HTTP methods such as GET or POST.
```

```python
# Q3. Why is redirect() used in Flask?
# Ans 3 =Flask class has a redirect() function. When called, it returns a response object and redirects
# the user to another target location with specified status code.
```

```python
# Q4. What are templates in Flask? Why is the render_template() function used?
# Ans 4 Templates in Flask are used to separate the presentation logic from the application logic. 
# They are HTML files that can be dynamically
# generated by Flask with the help of a templating engine.
# The render_template() function is used to render a template and return the HTML content to the client.
# It takes the name of the template file and any data that should be passed to the template as keyword arguments.
```
